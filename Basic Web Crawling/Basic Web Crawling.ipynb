### __Basic Web Crawling__

Traverse all the URLs in the page with max traversal depth of 3

__Submitted By:__ Auradee Castro

```python
import requests
import time
import random

from bs4 import BeautifulSoup
from urllib.parse import urlparse
```

```python
MAX_TRAVERSE_DEPTH = 3
MAX_URL_PER_PAGE = 10
urls_list = []

def traverseURLS(url, depth):

    print(f"DEPTH {depth} : {url}")

    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    # Get all anchor elements on the page with no duplicates
    anchor_elem_list = set(soup.find_all('a'))

    valid_url_ctr = 0
    for elem in anchor_elem_list:
        scraped_url = elem.get('href')
        result = urlparse(scraped_url)
        
        # Check if valid URL and the traversal depth has not reached the maximum value
        if result.scheme and result.netloc and depth < MAX_TRAVERSE_DEPTH:
            urls_list.append([scraped_url, (depth + 1)])
            traverseURLS(scraped_url, (depth + 1))
            valid_url_ctr = valid_url_ctr + 1
        
        # Get only the first 10 URLs on the page
        if valid_url_ctr == MAX_URL_PER_PAGE:
            break
    
    # Add delay before processing next URL
    time.sleep(random.randint(0, 5))


if __name__ == "__main__":
    url = 'https://medium.com/analytics-vidhya/what-is-a-confusion-matrix-d1c0f8feda5'
    urls_list.append([url, 0])
    traverseURLS(urls_list[0][0], urls_list[0][1])
```
